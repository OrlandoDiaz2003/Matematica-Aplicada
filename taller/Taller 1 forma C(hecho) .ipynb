4. En cuanto tiempo tendran que venir los equipos de emergencia

```python

import numpy as np
from scipy.optimize import fsolve

#Funcion Valparaiso
def V(t):
    return 0.9*t**2 + 12.5*t + 36 -100

#Funcion Santiago
def S(t):
    return 0.95*t**2 + 16.8*t + 31 -100

#Funcion concepcion
def C(t):
    return 1.1*t**2 + 19.4*t + 26 -100

t = np.linspace(0,100,1)

solucion_valpa = fsolve(V, t)
solucion_santi = fsolve(S,t)
solucion_conce = fsolve(C,t)
print("Los equipos de emergencia tendran que ir en")
print(f"Valparaiso: {solucion_valpa[0]:.2f} horas")
print(f"Santiago: {solucion_santi[0]:.2f} horas")
print(f"Concepcion: {solucion_conce[0]:.2f} horas")
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Funcion Valparaiso
def V(t):
    return 0.9*t**2 + 12.5*t + 36

#Funcion Santiago
def S(t):
    return 0.95*t**2 + 16.8*t + 31

#Funcion concepcion
def C(t):
    return 1.1*t**2 + 19.4*t + 26

print(f"La acumulacion de agua durante las primeras 6 horas en cada ciudad sera de \nValparaiso: {V(6):.2f} \nSantiago: {S(6):.2f} mm  \nConcepcion: {C(6)} \n")
print(f"La acumulacion de agua durante las ultimas 6 horas en cada ciudad sera de \nValparaiso: {V(12) - V(6):.2f} \nSantiago: {S(12) - S(6):.2f} mm  \nConcepcion: {C(12) - C(6):.2f}")
print("\nEn las ultimas 6 horas del dia la acumlacion de agua es superior a la de las primeras 6 por lo que el equipo 2 tendra que ir mas veces")
```
